# Like Posts.

```python
  
count=0   
while count<30:
    #getting the list of all posts and clicking one by one.
    post_list=driver.find_elements_by_xpath('//div[starts-with(@class,"Nnq7C")]/div[contains(@class,"kIKUG")]')
    time.sleep(1)
    post_list[count].click()
    
    #like each post by checking its stauts weather liked or not.
    time.sleep(2)
    like_btn=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='eo2As ']/section[@class='ltpMr Slqrh']/span/button")))
    like_unlike=BeautifulSoup(like_btn.get_attribute('innerHTML')).find('svg')['aria-label']
    
    #printing mssg corrosponding to the post status.
    if like_unlike=='Like':
        like_btn.click()
        print('liked'+' '+str(count+1))
    else:
        print('Already Liked'+' '+str(count+1))
    
    # clicking on cross button
    time.sleep(1)
    back = driver.find_element_by_xpath("//div[@class='_2dDPU CkGkG']/div[3]/button")
    back.click()
        
    time.sleep(1)
    
    #scrolling the post lists to get more post.
    if count==len(post_list)-1 and count!=30:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(1)
    count+=1
    
    #breaking if post==30.
    if count==30:
        print('Liked 30 posts')
        break
    
        
```

# followers of ‘foodtalkindia’.

```python
#QUESTION---6(1)
#Finding the search box.
search=driver.find_element_by_xpath('//input[contains(@class,"x3qfX")]')

#clearing the previous content.
search.clear()

#searching for 'foodtalkindia'.
search.send_keys('foodtalkindia')

#Making an explicit wait till the required profile gets loaded.
profile_link=WebDriverWait(driver,15).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a/div[@class="z556c"]')))

#clicking on the required profile.
profile_link.click()

time.sleep(5)
#clicking over the followers to extract the followers of 'foodtalkindia'
follower_btn=driver.find_element_by_partial_link_text('followers')
follower_btn.click()


#creating list to store the followers whom I follow on 'foodtalkindia'
name_of_follower=[]
count=0
while True:
    try:
        #explicit wait till the list of followers gets loaded.
        user_list=WebDriverWait(driver,15).until(EC.presence_of_all_elements_located((By.XPATH,'//a[contains(@class,"notranslate")]')))

        #if user_list i.e, list of followers <500 scroll the list.
        if(len(user_list)<500):
            element=WebDriverWait(driver,15).until(EC.presence_of_element_located((By.XPATH,'//ul[starts-with(@class,"jSC57")]')))
            driver.execute_script("arguments[0].scrollIntoView(0,10);", element)
            time.sleep(2)
        
        #addng each follower to name_of_follower.
        name_of_follower.append(user_list[count].text)
        count+=1
        
        #if 500 followers added break
        if count==500:
            break
            
    #handling the error of element is no longer attached to the DOM.
    except StaleElementReferenceException:
        continue
        
#closing the followers window.
time.sleep(5)
exit = driver.find_element_by_xpath('//div[contains(@class, "eiUFA")]/div/button')
exit.click()
print(name_of_follower)
```

# print all the followers of “foodtalkindia” that I am follow but those who don’t follow me.

# APPROACH:

1.) I have extarcted the list of my followers and the people I am following.
2.) I have extracted the followers of 'foodtalkindia' to whom I follow.
3.) followers who don't follow me= followers_I_follow -my_followers.

```python
#QUESTION---6(2).
#extracting my followers

#finding my profile icon and clicking the same.
my_profile=driver.find_element_by_xpath('//div[@class="_47KiJ"]/div[5]/a')
my_profile.click()

time.sleep(2)

#finding and clicking the follower btn to get the list of followers.
follower_btn=driver.find_element_by_partial_link_text('followers')
follower_btn.click()

#explict wait to load the followers list.
WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,'PZuss')))

#getting the current height of scrollwindow.
scroll_window_height=driver.execute_script('return document.querySelector(".PZuss").offsetHeight;')

#creating list to store the my followers.
my_followers=[]

#loading the follower window to its full height.
while True:
    driver.execute_script('arguments[0].scrollIntoView(0,arguments[1]);',driver.find_element_by_class_name('PZuss'),scroll_window_height)
    time.sleep(3)
    
    #getting the new height of scroll window.
    new_height=driver.execute_script('return document.querySelector(".PZuss").offsetHeight;')
    
    if scroll_window_height==new_height:
        break
    scroll_window_height=new_height
    
#Since the entire window is loaded, adding the name of users to my_followers list.
while True:
    try:
        user_list=driver.find_elements_by_xpath('//a[contains(@class,"notranslate")]')
        for i in user_list:
            my_followers.append(i.text)
        break
        
    #handling the error of element is no longer attached to the DOM.
    except StaleElementReferenceException:
        continue
        
time.sleep(2)

#closing the followers window.
exit = driver.find_element_by_xpath('//div[contains(@class, "eiUFA")]/div/button')
exit.click()
```

```python
#extracting my following list.

#finding and clicking the following btn to get the list of whom I follow.
following_btn=driver.find_element_by_partial_link_text('following')
following_btn.click()

#explict wait to load the following list.
WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,'PZuss')))

#getting the current height of scrollwindow.
scroll_window_height=driver.execute_script('return document.querySelector(".PZuss").offsetHeight;')

#creating list to store the my followings.
my_following=[]

#loading the following window to its full height.
while True:
    
    driver.execute_script('arguments[0].scrollIntoView(0,arguments[1]);',driver.find_element_by_class_name('PZuss'),scroll_window_height)
    time.sleep(3)
    
    #getting the new height of scroll window.
    new_height=driver.execute_script('return document.querySelector(".PZuss").offsetHeight;')
    if scroll_window_height==new_height:
        break
    scroll_window_height=new_height
    
#Since the entire window is loaded, adding the name of users to my_following list.
while True:
    try:
        user_list=driver.find_elements_by_xpath('//a[contains(@class,"notranslate")]')
        for i in user_list:
            my_following.append(i.text)
        break
    
    #handling the error of element is no longer attached to the DOM.
    except StaleElementReferenceException:
        continue

#closing the following window.
time.sleep(2)
exit = driver.find_element_by_xpath('//div[contains(@class, "eiUFA")]/div/button')
exit.click()
```

```python
#followers of foodtalkindia whom I follow.

#Finding the search box.
search=driver.find_element_by_xpath('//input[contains(@class,"x3qfX")]')

#clearing the previous content.
search.clear()

#searching for 'foodtalkindia'
search.send_keys('foodtalkindia')

#Making an explicit wait till the required profile gets loaded.
profile_link=WebDriverWait(driver,15).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a/div[@class="z556c"]')))

#clicking on the required profile.
profile_link.click()
time.sleep(2)

#explicit wait to load the element.
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME ,'zwlfE')))

#clicking the followed by link.
followed_by = driver.find_element_by_partial_link_text("Followed by")
followed_by.click()

#explicit wait for the followed by list.
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME ,'SfPzb')))

#clicking to view whole list.
see_all_followers = driver.find_element_by_partial_link_text("See All Followers")
see_all_followers.click()

#creating list to store the followers list
follower_whom_i_follow=[]

while True:
    #using exception staleelementreferenceexception
    try:
        #creating a list in which the usernames is to be stored
        foodtalkindia_followed_by_me = []
        
        # mainting count
        count=0
        while True:
            # locating the usernames of followers
            u_element = driver.find_elements_by_xpath('//a[contains(@class, "notranslate")]')
            
            #checking if we got the first 500 or not
            if len(u_element)<len(my_following):
                
                #exlicitly waiting until the list is loaded completly
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
                
                #scrolling the window
                driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
                time.sleep(1)
                
            #getting the top username on the window
            i=u_element[count]
            #appending the username to the sodelhi list
            foodtalkindia_followed_by_me.append(i.get_attribute('innerHTML'))
            if len(foodtalkindia_followed_by_me)>=len(my_following):
                break
            count+=1
        break
    except StaleElementReferenceException:
        continue

    
#closing the following window.  
time.sleep(5)
exit = driver.find_element_by_xpath('//div[contains(@class, "eiUFA")]/div/button')
exit.click()
```

```python
#to get the list of followers who don't follow me.

#creating set for all lists.
my_following=set(my_following)
follower_whom_i_follow=set(foodtalkindia_followed_by_me)
my_followers=set(my_followers)

#set difference between follower_whom_i_follow and my_followers.
followers_not_following_me=follower_whom_i_follow-my_followers

#printing the each profile of followers_not_following_me.
for i in followers_not_following_me:
    print(i)
```
